```python
import pandas as pd
from glob import glob
from sqlalchemy import create_engine
```

```python
room_path = "../crawling/RoomData/"
room_files = glob(room_path + "*.xlsx")

room_files
```

```python
room_df_list = [pd.read_excel(i).iloc[:, 1:] for i in room_files]
```

```python
df = pd.concat(room_df_list)
df.index = [i for i in range(len(df))]
df
```

```python
df.columns
```

## pd.DataFrame -> data base

```python
df.columns = ["lecture_number", 
              "language", 
              "subject_name", 
              "teacher", 
              "tearm", 
              "day_period", 
              "aim",
              "link", 
              "room"]
```

```python
engine = create_engine('sqlite:///test.sqlite3', 
                       echo=False,
                       encoding="utf-8")
df.to_sql("df", engine, if_exists="replace")
```

## data base -> pd.DataFrame

```python
query = """
select * from df
"""

df_2 = pd.read_sql(sql=query, con=engine)
```
